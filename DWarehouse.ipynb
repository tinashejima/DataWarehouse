```python
import os
import pandas as pd
import numpy as np
from pyhive import hive
import csv


host = '197.221.242.150'        
port =  10348
database = 'default'  

conn = hive.Connection(host=host, port=port, database=database)

cursor = conn.cursor()

cursor.execute('SELECT * FROM observation')

df = pd.DataFrame(cursor.fetchall())

# Get the column names from the cursor description
columns = [desc[0] for desc in cursor.description]

# Create a pandas DataFrame from the results
df = pd.DataFrame(df, columns=columns)

# Specify the directory and file name for the CSV file
directory = '/home/tynash/Desktop/Exported'
file_name = 'output.csv'


# Construct the complete file path
csv_file_path = os.path.join(directory, file_name)

# Save the DataFrame as a CSV file
#df.to_csv(csv_file_path, index=False, na_rep='')

# Open the CSV file in write mode
with open(csv_file_path, 'w', newline='') as csvfile:
    # Create a CSV writer object
    writer = csv.writer(csvfile)

    # Write the column names as the header
    writer.writerow(columns)

    # Fetch and write the data rows
    for row in cursor.fetchall():
        writer.writerow(row)




#print(df.head(50))
print(df) 


cursor.close()
conn.close()
```
